## Capturing NAN values with a new feature
It works well if the data are not missing completely at random or we can say that It works well if the data are **MNAR(Missing data not at Random).**

```python
import pandas as pd
import numpy as np
import matplotlib as plt
%matplotlib inline
```

```python
df=pd.read_csv('titanic.csv', usecols=['Age','Fare','Survived'])
df.head()
```

```python
df['Age_NAN']=np.where(df['Age'].isnull(),1,0)
```

```python
df
```

```python
median = df.Age.median()
median
```

```python
df['Age'].fillna(df.Age.median(),inplace=True)
```
